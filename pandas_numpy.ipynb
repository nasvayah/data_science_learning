### query()

```python
sp = students_perfomance.copy(deep=True)
```

```python
sp = sp \
    .rename(columns = 
            {'parental level of education' : 'parental_level_of_education',
             'test preparation course' : 'test_preparation_course',
             'math score' : 'math_score',	
             'reading score' : 'reading_score',
             'writing score' : 'writing_score'})        # переименовали колонки с пробелами
```

```python
sp.query("writing_score > 50 & gender == 'female'")
```

### filter()

```python
sp.filter(like='score')  ## взяли те колонки где есть score (по умолчанию для столбцов axis = 1)
```

```python
sp.filter(like='0', axis=0)  ##уже для индексов строк
```

```python
sp.filter(items=['gender', 'lunch'])  ## еще один параметр, которй просто отбирает по точному названию
```

# Группировка и агрегация

```python
sp.groupby('gender').mean(numeric_only=True)  #группировать можно по функции сразу
```

```python
sp.groupby('gender').aggregate({'math_score' : 'mean', 'reading_score' : 'mean'})
```

```python
## если мы не хотим чтобы индекс был теперь male female, а не 0 и 1 то
sp.groupby('gender', as_index=False) \
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns = {'math_score' : 'm_mean',
                       'reading_score' : 'r_mean'})
# ну и заодно переименовали колонки
```

```python
sp.groupby('gender').math_score.nunique() 
```

```python
sp.sort_values(['gender','math_score'], ascending=False).groupby('gender').head(5)
```

```python
# создадим новую колонку разными способами
sp['total_score'] = sp['math_score'] + sp['reading_score'] + sp['writing_score']
sp = sp.assign(total_score_log = np.log(sp.total_score))
sp
```

```python
my_data = pd.DataFrame({'type' : ['A', 'A', 'B', 'B'], 'value' : [10, 14, 12, 23]})
my_data
```

```python
my_stat = pd.read_csv('my_stat.csv')
subset_1 = my_stat.query(" (V1 > 0) & (V3 == 'A')")
subset_2 = my_stat.query(" (V2 != 10) & (V4 >= 1)")
subset_2


```

```python
my_stat['V5'] = my_stat['V1'] + my_stat['V4']
my_stat['V6'] = np.log(my_stat['V2'])
my_stat
```

```python
my_stat = my_stat.rename(columns = \
                {'V1' : 'session_value', 
                 'V2' : 'group',
                 'V3' : 'time',
                 'V4' : 'n_users'}).copy(deep=True)
```

```python
my_stat['session_value'] = my_stat['session_value'].fillna(0)

med = float(my_stat[my_stat['n_users'] > 0].n_users.median())
```

```python
my_stat['n_users'] = [med if user < 0 else user for user in my_stat['n_users']]
my_stat
```
